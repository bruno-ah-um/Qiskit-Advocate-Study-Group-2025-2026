### Example 2

```python
#Register Definition and naming
qr_a = QuantumRegister(2,'qreg1') # (2,name='qreg1')
qr_b = QuantumRegister(2,'qreg2') # (2,name='qreg2')
cr = ClassicalRegister(4, name= 'cr')
#define the quantum circuit
qc = QuantumCircuit(qr_a,qr_b,cr)
qc.h(qr_a) # qc.h(qr_a[0:2]); qc.h(qr_a[0]) ; qc.h(qr_a[1])
qc.x(qr_b)
qc.cx(qr_a[1], qr_b[0])
qc.draw('mpl') 
```

```python
qc = QuantumCircuit(3) # no classical Registers are pre-defined
qc.x([0,1,2]) #qc.x(range(3)) 
qc.measure_all()
a = r"measure_all() adds the appropriate registers automatically and names them 'meas'"
print(a)
print('The pre-defined classical registers are not used now')
qc.draw('mpl')
```

## Bell States:

```python
from IPython.display import display, Markdown, Latex
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, array_to_latex

# First Bell State
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)

display(Markdown("1st Bell State: $\phi$\u207A = ${1}\over{\sqrt{2}}$$ \cdot (|00> + |11>)$"))
display(array_to_latex(Statevector(qc)))

qc.draw('mpl')
```

```python
# Second Bell State
qc_a = qc.copy()
qc_a.z(0)

display(Markdown("2nd Bell State: $\phi$\u207B = ${1}\over{\sqrt{2}}$$ \cdot (|00> - |11>)$"))
display(array_to_latex(Statevector(qc_a)))

qc_a.draw('mpl')
```

```python
# Third Bell State
qc_b = qc.copy()
qc_b.x(0)

display(Markdown("3rd Bell State: $\psi$\u207A = ${1}\over{\sqrt{2}}$$ \cdot (|01> + |10>)$"))
display(array_to_latex(Statevector(qc_b)))

qc_b.draw('mpl')
```

```python
# Firth Bell State
qc_c = qc.copy()
qc_c.z(0)
qc_c.x(0)

display(Markdown("4th Bell State: $\psi$\u207B = ${1}\over{\sqrt{2}}$$ \cdot (|01> - |10>)$"))
display(array_to_latex(Statevector(qc_c)))

qc_c.draw('mpl')
```
